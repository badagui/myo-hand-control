## Functions

```python
def get_waveformlength(series):
    return abs(series.diff()).mean()

# optimized
def df_window_apply(df, window_ms, shift_ms, func):
    freq = 100
    window_size = int((window_ms * freq) / 1000)
    shift_size = int((shift_ms * freq) / 1000)
    
    n_samples = df.shape[0]
    n_shifts = int((n_samples-window_size)/shift_size)
    
    result_df = pd.DataFrame()
    
    if n_samples > window_size:
        actual_sample = window_size
        features = df.iloc[0:actual_sample, :].apply(func, raw=False)
        result_df = result_df.append(features, ignore_index=True)
        for _ in range(n_shifts):
            features = df.iloc[actual_sample:(actual_sample+shift_size), :].apply(func, raw=False)
            result_df = result_df.append(features, ignore_index=True)
            actual_sample = actual_sample + shift_size            
    return result_df

def define_vertical_marks(series):
    marks_list = list()
    for i in range(1, len(series)):
        if series[i] != series[i-1]: 
            if series[i] != 0:
                marks_list.append((i, series[i]))
            else:
                marks_list.append((i, series[i-1]))
    return marks_list

def benchmark_file(model, file, bench_start, bench_stop, plot_start=0, plot_stop=1000):
    
    FREQ_HZ = 100
    
    # start/stop sample
    start_s = bench_start * FREQ_HZ
    stop_s = bench_stop * FREQ_HZ
    
    # Mean Absolute Value
    #a = df_window_apply(file['emg'].loc[start_s:stop_s], 250, 100, np.mean)
    a = np.apply_along_axis(np.mean, 0, file['emg'][start_s:stop_s])
    
    # Waveform Length
    #b = df_window_apply(file['emg'].loc[start_s:stop_s], 250, 100, get_waveformlength)
    b = np.apply_along_axis(get_waveformlength, 0, file['emg'][start_s:stop_s])

    # join
    #features = pd.concat([a, b], ignore_index=True, axis=1)
    features = np.concatenate((a,b))
    features = features.reshape(1,-1)
    
    features = scaler.transform(features)

    y_pred = model.predict(features)
    
    
    restimulus_marks = define_vertical_marks(file['restimulus'][0])
    exercise = file['exercise'].squeeze()

    if exercise == 1:
        offset = 0
    if exercise == 2:
        offset = 12
    if exercise == 3:
        offset = 12 + 17

    # DB1 PLOT ANALYSYS
    START_s = plot_start
    STOP_s = plot_stop
    
    plot_marks = restimulus_marks

    width = (plot_stop - plot_start)/6
    if width < 5:
        width = 5
    #print('width', width)
    fig, ax_lst = plt.subplots(10, 1, figsize=(width, 12), dpi=120, sharex=True)  # a figure with a grid of Axes
    x = (file['emg'][0]).index.tolist()
    x = [x/FREQ_HZ for x in x]

    for ch in range(10):            
        # plot original data
        ax_lst[ch].plot(x, file['emg'][ch], linewidth=0.6)
        ax_lst[ch].set_ylim(0, 1)
        ax_lst[ch].set_xlim(START_s, STOP_s)
        ax_lst[ch].set_ylabel('CH{}'.format(ch+1))

        mark_colors = ['green', 'red', 'cyan', 'magenta', 'yellow', 'brown', 'gold', 'cadetblue', 'lightgreen', 'orange', 'tomato', 'darkcyan', 'goldenrod', 'pink', 'slategray', 'navy', 'sandybrown', 'firebrick', 'olive', 'salmon', 'peru']

        # benchmark bars
        #if ch == 0:
        if 1:
            for i in range(y_pred.size):
                if y_pred[i] == 0:
                    color_code = 'skyblue'
                else:
                    color_code = mark_colors[y_pred[i]%len(mark_colors)]
                ax_lst[ch].axvline(x=(i*10+12.5+start_s)/FREQ_HZ, color=color_code, linewidth=0.3)

        # vertical markings
        if plot_marks:
            for ind, code in plot_marks:
                if (code+offset) not in [13, 18, 19, 25, 26, 30, 43, 46, 47, 50, 51]:
                    ax_lst[ch].axvline(x=ind/FREQ_HZ, color='black', linewidth=2)
                else:
                    ax_lst[ch].axvline(x=ind/FREQ_HZ, color=mark_colors[(code+offset)%len(mark_colors)], linewidth=2)

    ax_lst[-1].set_xlabel('tempo (s)');

    subject_id = file['subject'].squeeze()
    subject_exercise = file['exercise'].squeeze()

    fig.suptitle("EMG - Indivíduo " + str(subject_id) + " Exercício " + str(subject_exercise));
    plt.tight_layout(rect=[0, 0, 1, .97])
    
    print(y_pred)
    
    #fig.savefig("bench")
    #plt.close(fig)
```
